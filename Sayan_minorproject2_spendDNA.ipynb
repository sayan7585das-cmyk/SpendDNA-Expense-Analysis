```python
def get_category(description):

    description = str(description).upper()

    # Food Delivery
    if "SWIGGY" in description or "ZOMATO" in description:
        return "Food Delivery"

    # Quick Commerce
    elif ("ZEPTO" in description or
          "BLINKIT" in description or
          "INSTAMART" in description or
          "GROFERS" in description):
        return "Quick Commerce"

    # Shopping
    elif ("AMAZON" in description or
          "AMZN" in description or
          "FKART" in description or
          "FLIPKART" in description or
          "MYNTRA" in description or
          "NYKAA" in description):
        return "Shopping"

    # Groceries
    elif ("DMART" in description or
          "BIGBASKET" in description or
          "SUPERMART" in description or
          "KIRANAKART" in description or
          "INNOVATIVE RETAIL" in description):
        return "Groceries"

    # Cafe
    elif ("STARBUCKS" in description or
          "THIRD WAVE" in description or
          "THIRDWAVE" in description or
          "TWC" in description or
          "CCD" in description or
          "COFFEE DAY" in description):
        return "Cafe"

    # Restaurants
    elif ("RESTAURANT" in description or
          "MEGHANA" in description or
          "EMPIRE" in description or
          "TRUFFLES" in description or
          "DINEOUT" in description):
        return "Restaurants"

    # Transport
    elif ("RAPIDO" in description or
          "ROPPEN" in description or
          "UBER" in description or
          "OLA" in description or
          "BMTC" in description or
          "TUMMOC" in description):
        return "Transport"

    # Fuel
    elif ("INDIAN OIL" in description or
          "PETROL" in description or
          "HP" in description):
        return "Fuel"

    # Utilities
    elif ("BESCOM" in description or
          "ELEC" in description or
          "BWSSB" in description or
          "WATER BILL" in description):
        return "Utilities"

    # Investment
    elif ("GROWW" in description or
          "ZERODHA" in description):
        return "Investment"

    # Subscription
    elif ("PRIME" in description or
          "NETFLIX" in description or
          "SPOTIFY" in description or
          "HOTSTAR" in description or
          "DISNEY" in description):
        return "Subscription"

    # Salary
    elif "SALARY" in description:
        return "Salary"

    # Recharge
    elif ("AIRTEL" in description or
          "JIO" in description):
        return "Recharge"

    # Cash Withdrawal
    elif "ATM" in description:
        return "Cash Withdrawal"

    # Rent
    elif "RENT" in description:
        return "Rent"

    else:
        return "Others"
```

Executive Summary


```python
# Income and Expense Summary

total_income = df[df["Type"] == "Credit"]["Amount"].sum()

total_expense = df[df["Type"] == "Debit"]["Amount"].sum()

net_savings = total_income - total_expense

saving_rate = (net_savings / total_income) * 100
```

```python
print("="*65)
print("               EXECUTIVE SUMMARY")
print("="*65)

print(f"Total Transactions : {len(df)}")

print(f"Total Income       : ₹{total_income:,.2f}")

print(f"Total Expense      : ₹{total_expense:,.2f}")

print(f"Net Savings        : ₹{net_savings:,.2f}")

print(f"Savings Rate       : {saving_rate:.2f}%")
```

Spending Categories

```python
expense = df[df["Type"] == "Debit"]

category_spending = (
    expense
    .groupby("Category")["Amount"]
    .sum()
    .sort_values(ascending=False)
)

top_categories = category_spending.head(5)
```

Monthly Spending Trend

```python
monthly_spending = (
    df[df["Type"] == "Debit"]
    .groupby("Month_Name")["Amount"]
    .sum()
)

# Display months in calendar order
month_order = [
    "January","February","March","April","May","June",
    "July","August","September","October","November","December"
]

monthly_spending = monthly_spending.reindex(month_order).dropna()

monthly_spending
```

```python
monthly_change = monthly_spending.pct_change() * 100

monthly_change = monthly_change.round(2)

monthly_change
```

```python
highest_month = monthly_spending.idxmax()
highest_amount = monthly_spending.max()

lowest_month = monthly_spending.idxmin()
lowest_amount = monthly_spending.min()
```

```python
print("="*65)
print("MONTHLY SPENDING TREND")
print("="*65)

for month, amount in monthly_spending.items():
    print(f"{month:<12} ₹{amount:,.2f}")

print("\nHighest Spending Month :", highest_month)
print(f"Amount : ₹{highest_amount:,.2f}")

print("\nLowest Spending Month :", lowest_month)
print(f"Amount : ₹{lowest_amount:,.2f}")
```

Time Pattern Analysis

```python
hourly_spending = (
    df[df["Type"] == "Debit"]
    .groupby("Hour")["Amount"]
    .sum()
    .sort_index()
)

hourly_spending
```

```python
peak_hour = hourly_spending.idxmax()
peak_amount = hourly_spending.max()

print("="*65)
print("HOURLY SPENDING")
print("="*65)

for hour, amount in hourly_spending.items():
    print(f"{hour:02d}:00 - ₹{amount:,.2f}")

print("\nPeak Spending Hour :", f"{peak_hour:02d}:00")
print(f"Amount : ₹{peak_amount:,.2f}")
```

```python
def get_time_period(hour):

    if 5 <= hour < 12:
        return "Morning"

    elif 12 <= hour < 17:
        return "Afternoon"

    elif 17 <= hour < 22:
        return "Evening"

    else:
        return "Night"
```

```python
df["Time_Period"] = df["Hour"].apply(get_time_period)
```

```python
time_period_spending = (
    df[df["Type"] == "Debit"]
    .groupby("Time_Period")["Amount"]
    .sum()
)

print("="*65)
print("SPENDING BY TIME OF DAY")
print("="*65)

for period, amount in time_period_spending.items():
    print(f"{period:<12} ₹{amount:,.2f}")
```

Anomaly Detection

```python
expense = df[df["Type"]=="Debit"].copy()

mean = expense["Amount"].mean()

std = expense["Amount"].std()

expense["Z_Score"] = (
    expense["Amount"] - mean
) / std
```

```python
anomalies = expense[expense["Z_Score"] > 3]

anomalies[
    ["Date","Description","Amount","Z_Score"]
]
```

```python
print("="*65)
print("ANOMALOUS TRANSACTIONS")
print("="*65)

print(f"Total Anomalies : {len(anomalies)}")

if len(anomalies) > 0:
    print("\nTop Anomalies\n")

    print(
        anomalies[
            ["Date","Description","Amount"]
        ].head(10)
    )
```

Spending Archetypes

```python
archetypes = []
```

```python
food_categories = ["Food Delivery", "Restaurants", "Cafe"]

food_spending = category_spending[
    category_spending.index.isin(food_categories)
].sum()

food_percentage = (food_spending / total_expense) * 100

if food_percentage >= 20:
    archetypes.append("🍔 Foodie")
```

```python
shopping_categories = ["Shopping", "Quick Commerce", "Groceries"]

shopping_spending = category_spending[
    category_spending.index.isin(shopping_categories)
].sum()

shopping_percentage = (shopping_spending / total_expense) * 100

if shopping_percentage >= 25:
    archetypes.append("🛍️ Shopaholic")
```

```python
investment = category_spending.get("Investment", 0)

if investment > 0:
    archetypes.append("📈 Investor")
```

```python
cafe = category_spending.get("Cafe", 0)

if cafe > 1000:
    archetypes.append("☕ Coffee Lover")
```

```python
night_spending = (
    df[
        (df["Type"]=="Debit") &
        (df["Time_Period"]=="Night")
    ]["Amount"].sum()
)

if night_spending > total_expense * 0.15:
    archetypes.append("🌙 Night Owl")
```

```python
print("="*65)
print("SPENDING ARCHETYPES")
print("="*65)

if archetypes:
    for item in archetypes:
        print(item)
else:
    print("No dominant spending archetype identified.")
```
